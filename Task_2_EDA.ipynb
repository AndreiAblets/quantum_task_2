```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
%matplotlib inline
```

```python
data=pd.read_csv('train.csv')
```

# Data types and missing values

```python
data.info()
```

# Statistical summary

```python
data.describe().transpose()
```

# Check the distribution of the target variable

```python
target = 'target'  
plt.figure(figsize=(10, 6))
sns.histplot(data[target], kde=True, bins=30)
plt.title('Distribution of Target Variable')
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.show()
```

# Correlation matrix

```python
# Calculate the correlation matrix
corr_matrix = data.corr()

# Extract correlations with the target variable
target_corr = corr_matrix[target].drop(target).sort_values(ascending=False)

# Display the top 10 positively and negatively correlated features
print("Top 10 positively correlated features:")
print(target_corr.head(10))
print("\nTop 10 negatively correlated features:")
print(target_corr.tail(10))
```

```python
# Visualize the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# Histograms of features

```python
data.hist(figsize=(20, 20), bins=30)
plt.show()
```

# Box plots to detect outliers

```python
selected_features = data.columns.drop([target])[:8]  # Select 8 features
plt.figure(figsize=(15, 10))
for i, col in enumerate(selected_features):
    plt.subplot(2, 4, i+1)
    sns.boxplot(y=data[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

# Relationship between features 6 and 8

```python
data.loc[:,['6','8']]
```

```python
feature_6 = data['6']
feature_8 = data['8']

# Calculate Pearson correlation coefficient
correlation = feature_6.corr(feature_8)
print(f"Correlation between Feature 6 and Feature 8: {correlation:.4f}")
```

```python
# Transform Feature 6: if value < 0, set to 0; else set to 1
feature_6_transformed = feature_6.apply(lambda x: 0 if x < 0 else 1)

# Check if the transformed Feature 6 matches Feature 8
features_match = feature_6_transformed.equals(feature_8)
print(f"Do transformed Feature 6 and Feature 8 match? {features_match}")
```

```python
feature_6[(feature_6_transformed!=feature_8)]
```

```python
feature_6[(feature_6_transformed!=feature_8)].value_counts()
```

Since the transformed feature 6 is identical to feature 8, feature 8 does not provide additional information. In this case, it's best to drop feature 8 to reduce redundancy.



```python
# Drop Feature 8 from the dataset
data.drop(columns=['8'], inplace=True)
print("Feature 8 dropped from the dataset to eliminate redundancy.")
```
